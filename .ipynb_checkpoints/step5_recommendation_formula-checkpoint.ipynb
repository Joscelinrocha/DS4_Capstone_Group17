```python
import pandas as pd
import os
import glob
import numpy as np
```

```python
##loading Twitter dataset

twitter_data = pd.read_csv('data/twitter_genre_after_LDA.csv')
twitter_data.head()
#twitter_data.shape
```

```python
##loading Spotify dataset

spotify_data = pd.read_csv('data/clean_spotify.csv')
spotify_data
```

```python
#Adding a new column mergeing the track name and the artist name

spotify_data["recommended_track"] = '"'+ spotify_data["track_name"] + '"' + " by " + spotify_data["artist_name"]
```

```python
spotify_data.head()
```

### Creating the code to randomly choose 3 songs

```python
twitter_data.head()
```

```python
import random
recs2 = []


for cell in twitter_data["genre"]:
    filtered_df = spotify_data[spotify_data['genre'] == cell].reset_index()
    length = len(filtered_df)
    random_list = random.sample(range(0, length), 3)
    recs = []
    recs2.append(recs)
    count = 0
    for i in random_list:
        count = count + 1
        recommendation = str(count) + ". " + filtered_df["recommended_track"][i] +"\n"
        recs.append(recommendation)

twitter_data["recommendation"] = recs2
```

```python
topic_label = {'punkrock':'Punk Rock', 'chill':'Chill', 'punk':'Punk',  'latino':'Latino', 
               'happy':'Happy', 'rock':'Rock','rnb':'RNB', 'edm':'EDM', 'goth':'Goth',
               'gospel':'Gospel', 'house':'House', 'songwriter':'Songwriter', 
               'comedy':'Comedy', 'anime':'Anime', 'Japanese Rock':'jrock', 'country':'Country', 
               'disco':'Disco', 'ambient':'Ambient', 'jazz':'Jazz', 'pop':'Pop', 'honkytonk':'Honkytonk',
               'New Age':'newage', 'heavymetal':'Heavy Metal', 'romance':'Romance', 'acoustic':'Acoustic', 
               'afrobeat':'Afrobeat', 'party':'Party', 'rockabilly':'Rockabilly', 'altrock':'Alternative Rock',
               'soul':'Soul', 'funk':'Funk', 'groove':'Groove', 'minimaltechno':'Minimal Techno', 
               'dancehall':'Dancehall',  'club':'Club', 'breakbeat':'Breakbeat', 'punkrock':'Punk Rock',
               'reggae':'Reggae'}

# mapping the dictionary with the dataframe to get the labels.
twitter_data['genre_formatted'] = twitter_data['genre'].map(topic_label)

formatted_recs = []
for cell in twitter_data["recommendation"]: 
    formatted = ''.join(cell)
    formatted_recs.append(formatted)
    
twitter_data['recommendation_as_str'] = formatted_recs
```
